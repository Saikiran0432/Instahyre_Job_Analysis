```python
#Converting categorical into numerical
def temp(val):
    if val == 1:
        return  '0-10'
    elif val == 10:
        return '10-50'
    elif val == 50:
        return '50-200'
    elif val == 200:
        return '200-500'  
    else:
        return '>1000'
```

```python
df = df[['job_id','job_title','job_location','Company_Id','Company','details_id','skills_requried','emp_count']]
```

```python

def jobs(val):
    val = val.split(' ')
    if 'Software' in val :
        return 'Software'
    elif 'Data' in val :
        return 'Data'
    elif 'Business' in val:
        return 'Business'
    elif 'Frontend' in val:
        return 'Web'
    elif 'Backend' in val:
        return 'Web'
    elif 'Stack' in val:
        return 'Web'
    elif 'DevOps' in val:
        return 'DevOps'
    elif 'Sales' in val:
        return 'Sales'
    elif 'Marketing' in val:
        return 'Marketing'
    elif 'Cloud' in val:
        return 'cloud'
    else:
        return 'other'
```

```python
df['industry'] = df['job_title'].map(jobs)
```
